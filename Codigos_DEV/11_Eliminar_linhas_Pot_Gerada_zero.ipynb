```python
import os
import pandas as pd
import numpy as np
import glob
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.stattools import ccf
```

```python
# Leitura do arquivo de segundos 
arq_comp = r'H:\Path_Python\Dados_Python_2024\Lameirinho\Ficheiros_PKL_Saida\Arquivos_concatenados\files_29_01_and_06_05_concatenados_sem_NaNs.pkl'
df_completo = pd.read_pickle(arq_comp)
```

```python
# Definir as linhas onde G2_Gerador_Potência é zero
zero_power_rows = df_completo[df_completo['G2_Gerador_Potência'] == 0].index

# Definir como NaN as variáveis G2_Gerador_Potência e outras variáveis nessas linhas
df_completo.loc[zero_power_rows] = np.nan

# Contar o número de linhas totalmente preenchidas com NaN
num_rows_with_all_nans = df_completo.isna().all(axis=1).sum()

print("Número de linhas totalmente preenchidas com NaN:", num_rows_with_all_nans)
```

```python
# Definir as linhas onde G2_Gerador_Potência é zero
zero_power_rows = df_completo[df_completo['G2_Gerador_Potência'] == 0].index

# Definir como NaN as variáveis G2_Gerador_Potência e outras variáveis nessas linhas
df_completo.loc[zero_power_rows] = np.nan

# Selecionar as primeiras 100000 linhas
df_primeiras_100000 = df_completo.head(1000000)

# Plotar a coluna G2_Gerador_Potência das primeiras 100000 linhas
plt.figure(figsize=(15, 7))
plt.plot(df_primeiras_100000.index, df_primeiras_100000['G2_Gerador_Potência'], label='G2_Gerador_Potência', color='b')

# Adicionar título e legendas
plt.title('Potência Gerada nas Primeiras 100.000 Linhas')
plt.xlabel('Index')
plt.ylabel('G2_Gerador_Potência')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
