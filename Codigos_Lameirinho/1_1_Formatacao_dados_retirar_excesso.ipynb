```python
import pandas as pd
import numpy as np
import glob
import pickle
```

```python
caminho_arquivo_pickle = r'H:\Path_Python\Dados_Python_2024\Lameirinho\dados_lameirinho_06_05_2024_original_sem_excesso.pkl'
df_parte = pd.read_pickle(caminho_arquivo_pickle)
```

```python
df_parte
```

```python
# Redefinir o índice do DataFrame
df_parte = df_parte.reset_index(drop=True)

df_parte
```

```python
# Criar uma nova coluna 'datetime' combinando 'Data' e 'Hora'
df_parte['datetime'] = pd.to_datetime(df_parte['Data'] + ' ' + df_parte['Hora'], format='%d/%m/%Y %H:%M:%S')

# Definir a data e hora alvo
data_hora_alvo = pd.to_datetime('06/05/2024 17:30:23', format='%d/%m/%Y %H:%M:%S')

# Encontrar o índice da linha que corresponde à data e hora alvo
linha_alvo = df_parte[df_parte['datetime'] == data_hora_alvo].index

if not linha_alvo.empty:
    numero_linha = linha_alvo[0]  # Obter o primeiro (e único) índice que corresponde
    print(f"O número da linha que tem a data e hora {data_hora_alvo} é: {numero_linha}")
else:
    print(f"Não foi encontrada nenhuma linha com a data e hora {data_hora_alvo}")

# # Opcional: Remover a coluna 'datetime' se não for mais necessária
# df_parte = df_parte.drop(columns=['datetime'])
```

```python
# Remover as linhas do índice 0 ao 599249
df_parte = df_parte.drop(index=range(0, 599250))
```

```python
# Redefinir o índice do DataFrame
df_parte = df_parte.reset_index(drop=True)

# Exibir as primeiras linhas do DataFrame resultante para verificar
print(df_parte.head())
```
