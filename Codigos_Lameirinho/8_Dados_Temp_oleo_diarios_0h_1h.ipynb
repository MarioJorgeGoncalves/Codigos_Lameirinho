```python
import os
import pandas as pd
import numpy as np
import glob
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Leitura do arquivo de segundos 
arq_comp = r'H:\Path_Python\Dados_Python_2024\Lameirinho\Ficheiros_PKL_Saida\Arquivos_concatenados\files_29_01_and_06_05_concatenados_sem_NaNs.pkl'
dfX = pd.read_pickle(arq_comp)
```

```python
# Converter o Ã­ndice para o tipo de dado DateTime
dfX.index = pd.to_datetime(dfX.index)

# Filtrar as linhas entre 00h00 e 01h00
df_filtered = dfX.between_time('00:00:00', '01:00:00')
```

```python
# Iterar sobre cada dia
for date in df_filtered.index.floor('D').unique():
    # Filtrar os dados para o dia especÃ­fico
    df_date = df_filtered[df_filtered.index.floor('D') == date]
    
    # Criar uma figura e eixo para o grÃ¡fico
    fig, ax = plt.subplots()
    
    # Plotar a temperatura do Ã³leo para o dia especÃ­fico
    ax.plot(df_date.index, df_date['G2_Oleo_Motor_Temp'], label='Temperatura do Ãleo')
    
    # ConfiguraÃ§Ãµes do eixo x
    ax.set_xlim(pd.to_datetime(date), pd.to_datetime(date) + pd.Timedelta(hours=1))
    ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=5))  # Intervalo de 5 minutos
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    
    # ConfiguraÃ§Ãµes do grÃ¡fico
    ax.set_xlabel('Hora')
    ax.set_ylabel('Temperatura do Ãleo')
    ax.set_title(f'Temperatura do Ãleo - {date}')
    ax.legend(loc='upper left')
    ax.grid(True)
    
    # Exibir o grÃ¡fico
    plt.show()
```
